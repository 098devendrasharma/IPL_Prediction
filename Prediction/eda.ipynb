```python
delivary_df = match_df.merge(df,on="match_id")
```

```python
delivary_df = delivary_df[delivary_df["innings"] == 2] 
```

```python
delivary_df["current_score"] = delivary_df.groupby("match_id")["runs_total_y"].cumsum()
```

```python
delivary_df["runs_left"] = delivary_df["runs_total_x"] - delivary_df["current_score"] 
```

```python
delivary_df["balls_left"] = 126 - ((delivary_df["over"]+1)*6 + delivary_df["ball"])
```

```python
delivary_df["wickets"] = 10 - delivary_df["team_wicket"]
```

```python
# current run rate = runs / overs
delivary_df["curr_run_rate"]= (delivary_df["current_score"]*6 / (120-delivary_df["balls_left"])).round(2)
```

```python
# required run rate
# If left_balls > 0, calculate RRR. Else, set it to 0.0
delivary_df["req_run_rate"] = np.where(
    delivary_df["balls_left"] > 0, 
    (delivary_df["runs_left"] * 6 / delivary_df["balls_left"]), 
    0
).round(2)
```

```python
# create the result for match if 1 ==> win batting team and 0 ==> loose match batting team 
def result(row):
    return 1 if row["batting_team"] == row["match_won_by"] else 0
```

```python
delivary_df["result"] = delivary_df.apply(result,axis = 1)
```

```python
final_df = delivary_df[['batting_team', 'bowling_team','city',"runs_left","balls_left","wickets","runs_total_x","curr_run_rate","req_run_rate","result"]]
final_df.shape
```

```python
# Suffle the dataset
final_df = final_df.sample(final_df.shape[0])
```

```python
final_df.isnull().sum()
```

```python
x = final_df.drop(columns=["result"])
y = final_df["result"] 
# train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size= 0.20, random_state=42)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
trf = ColumnTransformer([
    ('trf1', OneHotEncoder(sparse_output=False,drop="first"),["batting_team","bowling_team","city"]),
    # ('trf2',StandardScaler(),['runs_left', 'left_balls', 'current_run_rate', 'req_run_rate'])
     ],
    remainder="passthrough")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline

pipe = Pipeline(steps=[
    ('step1', trf),
    ('step2', LogisticRegression(solver='liblinear'))
])
```

```python
pipe.fit(x_train, y_train)
y_pred= pipe.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
import pickle

# The match_progression function logic for Win Probability Analysis
def match_progression(x_df, match_id, pipe):
    # Filter data for the specific match
    match = x_df[x_df['match_id'] == match_id]
    
    # We only care about the state at the end of each over (ball 6)
    match = match[(match['ball'] == 6)]
    
    # Select relevant features and drop missing values
    temp_df = match[[
    'batting_team', 
    'bowling_team', 
    'city', 
    'runs_left', 
    'balls_left',  # aapke index mein yehi naam hai
    'wickets', 
    'runs_total_x', # target score
    'curr_run_rate', 
    'req_run_rate'
        ]].dropna()
    # Avoid division by zero or end-of-game states
    temp_df = temp_df[temp_df['balls_left'] != 0]
    
    # Predict probabilities using the trained pipeline
    result = pipe.predict_proba(temp_df)
    
    # Add Win/Loss percentages to the dataframe
    temp_df['lose'] = np.round(result.T[0] * 100, 1)
    temp_df['win'] = np.round(result.T[1] * 100, 1)
    
    # Track the over numbers
    temp_df['end_of_over'] = range(1, temp_df.shape[0] + 1)
    
    # Calculate Runs Scored and Wickets Lost in each specific over
    target = temp_df['runs_total_x'].values[0]
    runs = list(temp_df['runs_left'].values)
    new_runs = runs[:]
    runs.insert(0, target)
    temp_df['runs_after_over'] = np.array(runs)[:-1] - np.array(new_runs)
    
    wickets = list(temp_df['wickets'].values)
    new_wickets = wickets[:]
    new_wickets.insert(0, 10)
    wickets.append(0)
    w = np.array(wickets)
    nw = np.array(new_wickets)
    temp_df['wickets_in_over'] = (nw - w)[0:temp_df.shape[0]]
    
    print("Target -", target)
    
    # Return the clean progression table and the target score
    temp_df = temp_df[['end_of_over','runs_after_over','wickets_in_over','lose','win']]
    return temp_df, target

```
